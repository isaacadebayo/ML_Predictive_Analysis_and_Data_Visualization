```python
import geopandas as gpd
import folium
import os
from google.colab import files

# Assuming files are already uploaded in a previous step, and 'uploaded' dict is available.
# If not, you'd need to run files.upload() here again.

# First, identify the base name of the SHP file that was actually used.
# From previous output, it was 'ne_110m_admin_0_countries (3).shp'
shp_file_name_uploaded = 'ne_110m_admin_0_countries (3).shp'
base_name = os.path.splitext(shp_file_name_uploaded)[0]

# List of expected extensions for a shapefile
extensions = ['.shp', '.dbf', '.shx', '.prj', '.cpg']

# Rename files to ensure consistent base name
for ext in extensions:
    original_name = base_name.replace(' (3)', '') + ext # Try original name without (3)
    new_name = base_name + ext # Target name with (3)

    if os.path.exists(original_name) and not os.path.exists(new_name):
        os.rename(original_name, new_name)
        print(f"Renamed '{original_name}' to '{new_name}'")

# Define the path to the corrected shapefile
global_shp_file_path = os.path.join(os.getcwd(), shp_file_name_uploaded)
print(f"Using shapefile path: {global_shp_file_path}")

# Load the shapefile with the corrected names
world_gdf = gpd.read_file(global_shp_file_path)

# --- DEBUG: Print columns to confirm correct country name field ---
print("\nColumns in world_gdf after loading with corrected filenames:")
print(world_gdf.columns)

# Aggregate longevity_df to get one life expectancy value per country
mean_life_expectancy_per_country = longevity_df.groupby('Country Name')['Life Expectancy World Bank'].mean().reset_index()

# Merge the world GeoDataFrame with your life expectancy data
# Assuming 'NAME' in world_gdf corresponds to 'Country Name' in your data
merged_gdf = world_gdf.merge(
    mean_life_expectancy_per_country,
    left_on='NAME', # Often 'NAME' or 'name_long' is the country name column in these shapefiles
    right_on='Country Name',
    how='left'
)

# --- DEBUG: Check merged data ---
print("\nMerged GeoDataFrame. Showing relevant columns for mapping:")
display(merged_gdf[['NAME', 'Country Name', 'Life Expectancy World Bank', 'geometry']].head())

# Create a Folium map
m = folium.Map(location=[0, 0], zoom_start=2)

# Add the choropleth layer
folium.Choropleth(
    geo_data=merged_gdf,
    name='choropleth',
    data=merged_gdf,
    columns=['NAME', 'Life Expectancy World Bank'],
    key_on='feature.properties.NAME',
    fill_color='YlGnBu',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Global Life Expectancy'
).add_to(m)

# Add layer control to toggle choropleth on/off
folium.LayerControl().add_to(m)

# Display the map
display(m)
```

```python
import geopandas as gpd
import folium
import os
import pandas as pd

# Assuming files are already uploaded in a previous step, and 'uploaded' dict is available.
# If not already defined, ensure global_shp_file_path and longevity_df are available.
# For demonstration, I'll re-initialize them using paths from previous execution.
shp_file_name_uploaded = 'ne_110m_admin_0_countries (3).shp'
global_shp_file_path = os.path.join(os.getcwd(), shp_file_name_uploaded)

world_gdf = gpd.read_file(global_shp_file_path)
longevity_df = pd.read_csv('longevity.csv') # Reload if necessary, or use the existing df from memory

# --- Concise Country Name Mapping and Data Aggregation ---
# Define a mapping for common country name discrepancies
country_name_mapping = {
    'United States': 'United States of America',
    'Congo, Dem. Rep.': 'Dem. Rep. Congo',
    'Congo, Rep.': 'Congo',
    'Egypt, Arab Rep.': 'Egypt',
    'Slovak Republic': 'Slovakia',
    'Micronesia, Fed. Sts.': 'Micronesia',
    'Kyrgyz Republic': 'Kyrgyzstan',
    'Yemen, Rep.': 'Yemen',
    'Cote d\'Ivoire': 'Ivory Coast',
    'Eswatini': 'Swaziland',
    'Timor-Leste': 'Timor-Leste',
    'Dominican Republic': 'Dominican Rep.',
    'Equatorial Guinea': 'Eq. Guinea',
    'North Macedonia': 'Macedonia',
    'Trinidad and Tobago': 'Trinidad and Tobago',
    'St. Vincent and Grenadines': 'St. Vincent and the Grenadines',
    'Venezuela, RB': 'Venezuela',
    'Bahamas, The': 'Bahamas',
    'Gambia, The': 'Gambia',
    'Cote d\'Ivoire': 'Côte d\'Ivoire',
    'Egypt': 'Egypt, Arab Rep.',
    #'S. Sudan': 'South Sudan',
    'South Sudan': 'S. Sudan',
    'Central African Republic': 'Central African Rep.',
    'Bosnia and Herzegovina': 'Bosnia and Herz.',
    'Czech Republic': 'Czechia',
    'Somaliland': 'Somalia',
    'Eswatini': 'eSwatini',
    #'eSwatini' : 'Eswatini',
    'North Macedonia': 'North Macedonia',


}

# Apply mapping to the 'Country Name' column of longevity_df
# Use .loc to modify the original DataFrame and create a new column if desired
longevity_df_corrected = longevity_df.copy()
longevity_df_corrected['Country Name Mapped'] = longevity_df_corrected['Country Name'].replace(country_name_mapping)

# Now, use the 'Country Name Mapped' for merging
mean_life_expectancy_per_country = longevity_df_corrected.groupby('Country Name Mapped')['Life Expectancy World Bank'].mean().reset_index()

# --- Determine Merge Column and Perform Merge ---
merge_column_name = None
for col in ['NAME', 'name_long', 'sovereign_name', 'admin', 'sovereignt']:
    if col in world_gdf.columns:
        merge_column_name = col
        break

if merge_column_name is None:
    print("Error: No suitable country name column found in GeoDataFrame for merging.")
else:
    merged_gdf_concise = world_gdf.merge(
        mean_life_expectancy_per_country,
        left_on=merge_column_name,
        right_on='Country Name Mapped',
        how='left'
    )

    # Identify countries that did not merge (Life Expectancy is NaN)
    unmatched_countries = merged_gdf_concise[merged_gdf_concise['Life Expectancy World Bank'].isna()]

    print("\nCountries in the map data (world_gdf) that did not find a match in your life expectancy data and are therefore greyed out:")
    # Sort by the merge_column_name for consistent output
    display(unmatched_countries[[merge_column_name, 'Country Name Mapped', 'Life Expectancy World Bank']].drop_duplicates().sort_values(by=merge_column_name))

    # Create a Folium map
    m = folium.Map(location=[0, 0], zoom_start=2)

    # Add the choropleth layer
    folium.Choropleth(
        geo_data=merged_gdf_concise,
        name='choropleth',
        data=merged_gdf_concise,
        columns=[merge_column_name, 'Life Expectancy World Bank'],
        key_on=f'feature.properties.{merge_column_name}',
        fill_color='YlGnBu',
        fill_opacity=0.7,
        line_opacity=0.2,
        legend_name='Global Life Expectancy'
    ).add_to(m)

    # Add layer control to toggle choropleth on/off
    folium.LayerControl().add_to(m)

    # Display the map
    display(m)
```
